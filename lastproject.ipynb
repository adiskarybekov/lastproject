```python
import pandas as pd
import matplotlib.pyplot as plt
import regex as re
import statistics
```

```python
pokedex = pd.read_csv('pokedex.csv')
pokedex.head()
```

Now I want to find how many pokemon have a mega evalution?


```python
mega_pokemon = []
pokemons = pokedex['name']
for pokemon in pokemons:
  if re.search('Mega ', pokemon):
    print(pokemon)
    mega_pokemon.append(pokemon)
print(len(mega_pokemon), "pokemons have a mega evolutions")
```
